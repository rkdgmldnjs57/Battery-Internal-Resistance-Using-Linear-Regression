# 물리학 실험 - 건전지 내부저항 값 선형 회귀

```python
import matplotlib.pyplot as plt
import random
import numpy as np
dataI = np.array([0.03, 0.07, 0.11, 0.15, 0.19, 0.23, 0.27, 0.31, 0.35, 0.39])
dataV = np.array([1.49, 1.48, 1.45, 1.42, 1.41, 1.40, 1.38, 1.35, 1.33, 1.30])

#V = e-Ir이지만 편의상 V = Ir + e로 설정

plt.scatter(dataI, dataV)
```

```python
e = 5.0
r = 5.0
sum = 0.0
length = len(dataI)
learning_rate = 0.001
epochs = 1000000
```

```python
#최소제곱법 Sum(e-IR-V)^2 / length
for cnt in range(epochs) : 
    
    cost = np.sum((e+dataI*r-dataV)**2)/length
    if(not cnt%50000) : 
        print(e, r, cost)
        x_label = np.array([0.0, 0.05, 0.10, 0.15, 0.20, 0.25, 0.30, 0.35, 0.40, 0.45, 0.50])
        y_label = x_label*r + e
        plt.scatter(dataI, dataV)
        plt.plot(x_label, y_label)
        plt.show()
    gradient_e = np.sum(2*(e+dataI*r-dataV))/length
    gradient_r = np.sum(2*dataI*(e+dataI*r-dataV))/length
    e -= gradient_e * learning_rate
    r -= gradient_r * learning_rate
    
```

```python
x_label = np.array([0.0, 0.05, 0.10, 0.15, 0.20, 0.25, 0.30, 0.35, 0.40, 0.45, 0.50])
y_label = x_label*r + e
plt.scatter(dataI, dataV)
plt.plot(x_label, y_label)
plt.show()
```
